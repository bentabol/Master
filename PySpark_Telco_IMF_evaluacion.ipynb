**9. Queremos obtener un Coeficiente de Ponderación que nos permita evaluar a cada cliente en función de su consumo para identificar los clientes más atractivos que forman parte de nuestra compañía. ESTE CÁLCULO SÓLO SE REALIZARÁ PARA LOS CLIENTES QUE TENGAN UNA SOLA OFERTA CONTRATADA CON LA COMPAÑÍA. </br> Este coeficiente se obtendrá en base a los consumos diarios, y por tanto sólo se tendrán en cuenta los clientes que existen en el mes de Agosto, ya que no tenemos datos de consumo del mes de Julio. </br> Las ponderaciones que se darán a cada uno de los consumos son las siguientes:** 
- 0.4 --> llamadas desde telefóno móvil.
- 0.3 --> datos_móviles_MB
- 0.2 --> llamadas desde teléfono fijo
- 0.1 --> SMS enviados </br> 

**Los pasos a seguir son los siguientes:**
  1. Obtener la suma de todos los días de cada uno de los 4 consumos para cada cliente.
  2. Obtener el máximo del cálculo anterior de todos los clientes para poder obtener un valor entre 0 y 1 para cada uno de los 4 consumos (recordar que sólo aplica para los clientes con UNA SOLA OFERTA CONTRATADA en el mes de Agosto. El cliente con mayor consumo de datos tendrá un valor de 1 en la columna 'datos_móviles_0_1')
  3. Multiplicar estas columnas obtenidas con valor entre 0 y 1 por su ponderación correspondiente (por ejemplo la columna "datos_moviles_0_1" se multiplicará por la ponderación de datos móviles 0.4).
  4. Por último se calculará la suma de las 4 ponderaciones obtenidas para calcular el Coeficiente de ponderación buscado (coeficiente_cliente), casteado a 3 decimales.
  5. Mostrar DF resultante ordenado por este coeficiente en sentido descendente, de manera que el primer registro corresponderá al cliente más atractivo. **El DF resultante tendrá 3 columnas: nombre, edad, coeficiente_cliente.**

```
# Importar las funciones necesarias
from pyspark.sql.functions import col, sum as _sum, max as _max, row_number

# Cargar las tablas de clientes y consumos diarios
df_clientes = spark.read.table("df_clientes")
df_facturas_mes_actual = spark.read.table("df_facturas_mes_actual")
df_consumos_diarios = spark.read.table("df_consumos_diarios")

# Filtrar solo los clientes con una sola oferta contratada
df_clientes_una_oferta = df_facturas_mes_actual.groupBy("id_cliente").agg(_max("id_oferta").alias("n_ofertas")) \
    .filter(col("n_ofertas") == 1)

# Unir los clientes con una sola oferta con los consumos diarios
df_consumo_una_oferta = df_consumos_diarios.join(df_clientes_una_oferta, "id_cliente")

# Agrupar por id_cliente para obtener el total de cada consumo
df_consumo_total = df_consumo_una_oferta.groupBy("id_cliente").agg(
    _sum("minutos_llamadas_movil").alias("total_minutos_movil"),
    _sum("consumo_datos_MB").alias("total_datos_moviles"),
    _sum("minutos_llamadas_fijo").alias("total_minutos_fijo"),
    _sum("sms_enviados").alias("total_sms_enviados")
)

# Obtener el máximo de cada tipo de consumo para normalizar
max_minutos_movil = df_consumo_total.agg(_max("total_minutos_movil")).collect()[0][0]
max_datos_moviles = df_consumo_total.agg(_max("total_datos_moviles")).collect()[0][0]
max_minutos_fijo = df_consumo_total.agg(_max("total_minutos_fijo")).collect()[0][0]
max_sms_enviados = df_consumo_total.agg(_max("total_sms_enviados")).collect()[0][0]

# Normalizar los consumos
df_normalizado = df_consumo_total.withColumn(
    "minutos_movil_0_1", col("total_minutos_movil") / max_minutos_movil
).withColumn(
    "datos_moviles_0_1", col("total_datos_moviles") / max_datos_moviles
).withColumn(
    "minutos_fijo_0_1", col("total_minutos_fijo") / max_minutos_fijo
).withColumn(
    "sms_enviados_0_1", col("total_sms_enviados") / max_sms_enviados
)

# Aplicar las ponderaciones para calcular el coeficiente de ponderación
df_ponderado = df_normalizado.withColumn(
    "coeficiente_cliente", 
    (col("minutos_movil_0_1") * 0.4 +
     col("datos_moviles_0_1") * 0.3 +
     col("minutos_fijo_0_1") * 0.2 +
     col("sms_enviados_0_1") * 0.1)
)

# Unir con la tabla de clientes para obtener el nombre y la edad
df_resultado = df_ponderado.join(df_clientes, "id_cliente").select(
    "nombre", "edad", "coeficiente_cliente"
)

# Mostrar el dataframe resultante ordenado por el coeficiente en sentido descendente
df_resultado.orderBy(col("coeficiente_cliente").desc()).show()
```

**10. Queremos averiguar la fecha en la que entre los 3 clientes que más datos consumen de cada grupo_edad llegan a un consumo de 20 GB de datos móviles. En caso de que algún grupo_edad no llegue entre los 3 clientes a 20 GB en todo el mes, se asignará "null" como valor de esta columna (Recordar que 1 GB = 1024 MB). </br> Obtener un DF que contiene 4 registros, uno para cada grupo_edad y 3 columnas: grupo_edad, fecha_20_GB, datos_moviles_total_grupo_3_clientes.** (datos_moviles_total_grupo_3_clientes representa el total de datos consumidos en MB por los 3 clientes del grupo hasta final de mes)

```
# Importar las funciones necesarias
from pyspark.sql.functions import col, sum as _sum, min as _min, when
from pyspark.sql.window import Window

# Cargar las tablas de clientes y consumos diarios
df_clientes = spark.read.table("df_clientes")
df_consumos_diarios = spark.read.table("df_consumos_diarios")

# Crear la columna "grupo_edad" con los rangos de edad
df_clientes = df_clientes.withColumn(
    "grupo_edad",
    when((col("edad") >= 18) & (col("edad") <= 25), 1)
    .when((col("edad") >= 26) & (col("edad") <= 40), 2)
    .when((col("edad") >= 41) & (col("edad") <= 65), 3)
    .otherwise(4)
)

# Unir los clientes con los consumos diarios
df_consumos_con_grupo = df_consumos_diarios.join(df_clientes, "id_cliente")

# Agrupar por grupo de edad, cliente, y fecha para sumar el consumo diario de datos móviles
df_consumo_diario_agrupado = df_consumos_con_grupo.groupBy("grupo_edad", "id_cliente", "fecha").agg(
    _sum("consumo_datos_MB").alias("datos_diarios_MB")
)

# Sumar el consumo total por cliente para identificar a los 3 clientes con mayor consumo en cada grupo de edad
windowSpec = Window.partitionBy("grupo_edad").orderBy(col("total_datos_mb").desc())

df_consumo_por_cliente = df_consumo_diario_agrupado.groupBy("grupo_edad", "id_cliente").agg(
    _sum("datos_diarios_MB").alias("total_datos_mb")
).withColumn("rank", row_number().over(windowSpec)).filter(col("rank") <= 3)

# Obtener el consumo diario acumulado de los 3 principales clientes por grupo_edad
df_top3_consumo = df_consumo_diario_agrupado.join(df_consumo_por_cliente, ["grupo_edad", "id_cliente"]).orderBy("grupo_edad", "fecha")

# Usar una ventana para calcular el consumo acumulado por grupo de edad y fecha
windowSpecFecha = Window.partitionBy("grupo_edad").orderBy("fecha").rowsBetween(Window.unboundedPreceding, Window.currentRow)

df_acumulado = df_top3_consumo.withColumn(
    "consumo_acumulado_MB", _sum("datos_diarios_MB").over(windowSpecFecha)
)

# Identificar la fecha en la que se alcanza o supera el umbral de 20 GB (20480 MB)
df_acumulado_20GB = df_acumulado.filter(col("consumo_acumulado_MB") >= 20480).groupBy("grupo_edad").agg(
    _min("fecha").alias("fecha_20_GB"),  # Obtener la primera fecha en la que se alcanza 20 GB
    _sum("consumo_acumulado_MB").alias("datos_moviles_total_grupo_3_clientes")
)

# Manejar los grupos que no alcanzan los 20 GB
df_consumo_total_por_grupo = df_top3_consumo.groupBy("grupo_edad").agg(
    _sum("datos_diarios_MB").alias("total_consumo_MB")
)

# Filtrar los grupos que no llegan a 20480 MB y asignarles null en la columna de fecha
df_grupos_no_20GB = df_consumo_total_por_grupo.filter(col("total_consumo_MB") < 20480).select(
    "grupo_edad", 
    when(col("total_consumo_MB") < 20480, None).alias("fecha_20_GB"),
    col("total_consumo_MB").alias("datos_moviles_total_grupo_3_clientes")
)

# Unir ambos resultados (los que alcanzan 20 GB y los que no)
df_resultado_final = df_acumulado_20GB.unionByName(df_grupos_no_20GB)

# Mostrar el dataframe final con las columnas solicitadas
df_resultado_final.orderBy("grupo_edad").show()
```
